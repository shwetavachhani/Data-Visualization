```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as mtick

df = pd.read_csv('population.csv')
```

# Quesion 1

 Recreate the following plot using Matplotlib’s functional API. This line plot shows the population of Canada between the years 1955 and 2020. Make sure to include the axes labels, title, and that the tick marks on the y-axis match the figure below.


```python
#References
# for gca() -> https://stackoverflow.com/questions/62785542/module-matplotlib-pyplot-has-no-attribute-yaxis
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axis.Axis.set_major_formatter.html


Record = df.loc[((df['Year'] >= 1950) & (df['Year'] <= 2020)) & (df['Country'] == 'Canada')]

plt.plot(Record['Year'], Record['Population'])

def format_number(data_value,indx):
    if data_value >= 1_000_000:
        formatter = '{:1.0f}M'.format(data_value*0.000_001)
    else:
        formatter = '{:1.0f}K'.format(data_value*0.001)
    return formatter

plt.xlim((1951,2022))
plt.yticks(np.arange(15000000,max(Record['Population']),5000000))
plt.gca().yaxis.set_major_formatter(format_number)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("Canada's Population Growth")
```

## Question-2

Recreate the following plot using Matplotlib’s functional API. This line plot shows the percentage of the world’s population that are living in urban areas. This has increased from low 30% to mid 50% over the past 60 years. 

You will need to use Pandas to compute the required data. Specifically, you will need to obtain the percentage of the global population that live in urban areas on a per year basis. Make sure to include the axes labels, title, and that the tick marks on the y-axis match the figure below.

```python
# Reference
# for FormatStrFormatter -> https://stackoverflow.com/questions/31357611/format-y-axis-as-percent

data = df.groupby(['Year'])['Continent', 'Country','UrbanPopulation','Population'].sum().reset_index()

df['PoulationPercentage'] = data['UrbanPopulation']/data['Population']*100

x = df['Year']
y = df['PoulationPercentage']

plt.plot(x, y)
plt.ylim((30, 59)) 

fmt = '%.0f%%' 
yticks = mtick.FormatStrFormatter(fmt)
plt.gca().yaxis.set_major_formatter(yticks)

plt.xlabel('Year')
plt.ylabel('Urban Perctange')
plt.title('Percentage of Global Population living in Urban Areas')

```

## Question-3

Recreate the following plot using Matplotlib’s object-oriented API. 

This line plot shows the yearly
population change (absolute number) per continent over the past 60 years. You will need to compute
this data on a per continent basis using the country values. 

Make sure to include the axes labels
and plot title, and that the colours of each line match the figure. The legend should be in the centre
of the plot without a border (hint: framealpha), and the continents in alphabetical order.

```python

fig = plt.figure()

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0])


data = df[(df['Continent'] == 'Africa') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="Africa")


data = df[(df['Continent'] == 'Asia') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="Asia")


data = df[(df['Continent'] == 'Europe') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="Europe")


data = df[(df['Continent'] == 'Latin America and the Caribbean') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="Latin America and the Caribbean")

data = df[(df['Continent'] == 'North America') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="North America")

data = df[(df['Continent'] == 'Oceania') & ( df['Year'] >= 1950)]
rec = data.groupby(['Year'])['Continent', 'Country','Yearly%Change','YearlyChange','Population','MedianAge','UrbanPopulation'].sum().reset_index()
ax.plot(rec['Year'], (rec['YearlyChange']),label="Oceania")

ax.set_ylim((-0.5e7, 65e6))
ax.set_xlabel('Year')
ax.set_ylabel('Yearly Population Change')
ax.set_title('Yearly Population Change by Continent')
ax.legend(loc='center',frameon=False)

```

## Question-4

Recreate the following plot using Matplotlib’s object-oriented API. This scatter plot shows the
median age and percentage population growth (Yearly%Change) from the year 2019 to 2020 for all
the countries of the world. Each country is colour coded according to its continent. Notice how
countries with younger median ages have larger population growth rates.
Make sure to include the axes labels and plot title, and that the colours match the figure. (The
colours are consistent with the plot from question 3 – this reinforces their interpretation.) Exclude
countries that are missing median age information. The legend should be in the lower left corner,
without a border, and the continents in alphabetical order.


```python
Record = df.loc[(df['Year'] > 2019) & (df['Year'] <= 2020)]
q4= Record.rename(columns={"Yearly%Change": "YearlyPerChange"})

fig = plt.figure()

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0])

ax.scatter(q4[q4.Continent == 'Africa'].MedianAge, q4[q4.Continent == 'Africa'].YearlyPerChange,c='#1F76B4',label='Africa')
ax.scatter(q4[q4.Continent == 'Asia'].MedianAge, q4[q4.Continent == 'Asia'].YearlyPerChange,c='#FF7F0F',label='Asia')
ax.scatter(q4[q4.Continent == 'Europe'].MedianAge, q4[q4.Continent == 'Europe'].YearlyPerChange,c='#2AA02B',label='Europe')
ax.scatter(q4[q4.Continent == 'Latin America and the Caribbean'].MedianAge, q4[q4.Continent == 'Latin America and the Caribbean'].YearlyPerChange,c='#D62727',label='Latin America and the Caribbean')
ax.scatter(q4[q4.Continent == 'North America'].MedianAge, q4[q4.Continent == 'North America'].YearlyPerChange,c='#9367BD',label='North America')
ax.scatter(q4[q4.Continent == 'Oceania'].MedianAge, q4[q4.Continent == 'Oceania'].YearlyPerChange,c='#8B554B',label="Oceania")

ax.set_xlim((13, 50))
ax.set_ylim((-2.7, 4.2))
ax.set_xlabel('Median Age')
ax.set_ylabel('Percent')
ax.set_title("Country Population Growths (2019 to 2020)")

ax.legend(loc='lower left',frameon=False)

```

### Question 5

Recreate the following plot using Matplotlib’s object-oriented API. This bar plot shows the population
of the top 10 most populous countries in 2020.
Make sure to include the axes labels and title, and that the colours and x-axis ticks match the figure.
The countries should be in descending order in according to their population

```python
# change format from this video https://www.youtube.com/watch?v=l-j9Mmi7VPk


df2 = df[(df['Year'] == 2020)].nlargest(10,'Population')

def format_number(data_value,indx):
    if data_value >= 1_000_000_000:
        formatter = '{:1.1f}B'.format(data_value*0.000_000_001)
    else:
        formatter = '{:1.0f}M'.format(data_value*0.000_001)
    return formatter


label=df2.Country
values=df2.Population

fig = plt.figure()

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0])

ax.barh(label,values,color=['#1F76B4', '#FF7F0F', '#2AA02B', '#D62727', '#9367BD','#8B554B', '#E377C2', '#7F7F7F', '#BCBD21', '#15BECF'])
ax.invert_yaxis()

ax.set_xticks(np.arange(0,max(values)+1,200000000))
ax.xaxis.set_major_formatter(format_number)
ax.set_title("Top 10 Countries by Population (2020)")


```
