```python
# groupby date with satisfaction count

data =ln.loc[(ln['start_date'] > '2022-05-01') & (ln['start_date'] <= '2022-05-10')  & (ln['job_type'] == 'basic_lawncare')]
data = data.groupby([ 'customer_id'])['customer_satisfaction'].max().reset_index()
x = data['customer_id']
y = data['customer_satisfaction']

fig = plt.figure()
ax = fig.add_axes([0.2,0.2,0.75,0.75]) # axis starts at 0.1, 0.1
ax.set_title("Higher Customer satisfaction rates")
ax.set_xlabel("Satisfaction Rates")
ax.set_ylabel("Cusomers")
ax.barh(x,y)
fig.savefig("with_max_rating")
```

```python
data1 =ln.loc[(ln['start_date'] > '2022-05-01') & (ln['start_date'] <= '2022-05-10') 
               & (ln['job_type'] == 'basic_lawncare')]

data1 = data1.groupby([ 'customer_id'])['customer_satisfaction'].mean().reset_index()
x = data1['customer_id']
y = data1['customer_satisfaction']

fig = plt.figure()
ax = fig.add_axes([0.2,0.2,0.75,0.75])
ax.set_title("Average of Customer satisfaction rates")
ax.set_xlabel("Satisfaction Rates")
ax.set_ylabel("Cusomers")
ax.barh(x,y)
fig.savefig("with_avg_rating")
```

```python
data2 =ln.loc[(ln['start_date'] > '2022-05-01') & (ln['start_date'] <= '2022-05-10') 
             ]
data2 = data2.groupby([ 'job_type'])['material_costs','invoice_amount'].sum().reset_index()

fig, ax = plt.subplots()
ax.scatter(data2[data2.job_type == 'basic_lawncare'].material_costs, data2[data2.job_type == 'basic_lawncare'].invoice_amount,c='#1F76B4',label="basic_lawncare")
ax.scatter(data2[data2.job_type == 'complete_lawncare'].material_costs, data2[data2.job_type == 'complete_lawncare'].invoice_amount,c='#FF7F0F',label="complete_lawncare")
ax.scatter(data2[data2.job_type == 'garden_landscaping'].material_costs, data2[data2.job_type == 'garden_landscaping'].invoice_amount,c='#080b0d',label="garden_landscaping")
ax.scatter(data2[data2.job_type == 'install_new_lawn'].material_costs, data2[data2.job_type == 'install_new_lawn'].invoice_amount,c='#D62727',label="install_new_lawn")
ax.scatter(data2[data2.job_type == 'retaining_wall'].material_costs, data2[data2.job_type == 'retaining_wall'].invoice_amount,c='#9367BD',label="retaining_wall")
ax.scatter(data2[data2.job_type == 'spring_cleanup'].material_costs, data2[data2.job_type == 'spring_cleanup'].invoice_amount,c='#D62753',label="spring_cleanup")
ax.scatter(data2[data2.job_type == 'tree_pruning'].material_costs, data2[data2.job_type == 'tree_pruning'].invoice_amount,c='#2AA123',label="tree_pruning")
ax.scatter(data2[data2.job_type == 'water_feature'].material_costs, data2[data2.job_type == 'water_feature'].invoice_amount,c='#8B554B',label="water_feature")

ax.legend(loc='lower right',frameon=False)

ax.set_title("Profit Calculation with Material Cost and Invoice per job Type ")
ax.set_xlabel("Material costs")
ax.set_ylabel("Invoice amount")
fig.savefig("Profit_Calculation")
```
