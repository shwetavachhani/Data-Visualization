```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from pandas import DataFrame

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display


diamonds = pd.read_csv('diamonds.csv')
names = pd.read_csv('names.csv')
diamonds
```

# Quesion 1

#### The following heatmap shows the average carat weight of diamonds relative to their clarity and colour. An interactive dropdown menu allows for the data to be filtered around the origin of the diamonds. The “Natural” option filters to only naturally occurring diamonds, “Lab” filters to manufactured diamonds, and “All’ applies no filtering. In the example below, we can see that the heaviest diamonds are those with the poorest quality.

```python
# idea https://stackoverflow.com/questions/61821652/using-drop-down-widget-from-ipywidgets

df1 = diamonds[['colour','clarity','carat','type']]
df1=df1.sort_values(by='carat',ascending=False)
categories = {'All':'all','Natural': 'natural', 'Lab': 'lab'}
@interact(category = widgets.Dropdown(options=categories, description='Origin:'))
def f(category):
    if category == 'natural':
        df2=df1[(df1['type'] == 'natural')]
    elif category == 'lab':
        df2=df1[(df1['type'] == 'lab')]
    else :
        df2=df1
    heatmap_pt = pd.pivot_table(df2,values ='carat', index=['colour'], columns=['clarity'])
    heatmap_pt = pd.DataFrame(heatmap_pt, columns =['SI2','SI1','VS2','VS1','VVS2','VVS1','IF'])
    plt.figure(figsize = (7,5))
    s=sns.heatmap(heatmap_pt, cmap='mako',),
    plt.xlabel('clarity'),
    plt.ylabel('color'),
    plt.show(block=False)
```

# Quesion 2

#### Re-create the interactive experience in the “Top baby names” visualization published by the UK Office for National Statistics, which can be found here:

```python
df2 = names.copy()
df2['name'] = df2['name'].str.strip()
df2_boys = df2[df2.sex == 'boy'].reset_index(drop=True)
df2_girls = df2[df2.sex == 'girl'].reset_index(drop=True)
df2_100_boy = df2_boys[df2_boys.year == 2020].sort_values(['rank'], ascending=True).head(100).reset_index(drop=True)
df2_100_girl = df2_girls[df2_girls.year == 2020].sort_values(['rank'], ascending=True).head(100).reset_index(drop=True)
```

```python
options = {'Baby boy names': 'boy', 'Baby girl names':'girl'}

def f(sex, filter):
    
    #filter names
    df2_ent = pd.DataFrame()
    df2_hgh = pd.DataFrame()
    df2_grey = pd.DataFrame()
    df2_ent['names'] = filter.split()[:6]
    for name in  df2_ent.names:
        df2_ent['names'] = df2_ent['names'].str.upper()
    
    #if girl filter girl records
    if(sex=='girl'):
        df2_grey = df2_girls[df2_girls.name.isin(df2_100_girl.name)].reset_index(drop=True)
        df2_hgh = df2_girls[df2_girls.name.isin(df2_ent.names) & df2_girls.name.isin(df2_100_girl.name)].reset_index(drop=True)
    # else boy records
    elif(sex=='boy'):
        df2_grey = df2_boys[df2_boys.name.isin(df2_100_boy.name)].reset_index(drop=True)
        df2_hgh = df2_boys[df2_boys.name.isin(df2_ent.names) & df2_boys.name.isin(df2_100_boy.name)].reset_index(drop=True)
     
    fig2, ax2 = plt.subplots(figsize=(13,9))
    #grey plot
    sns.lineplot(x = 'year', y = 'rank', data = df2_grey, hue = 'name', legend=False, 
                 palette=sns.color_palette(['gray'], 100), ax=ax2, alpha=0.7)

    #highlight
    if not (df2_hgh.empty):
        #sorting based on the entered order
        df2_hgh = df2_hgh.sort_values(by=['name'], key=lambda x: pd.Index(df2_ent['names']).get_indexer(x))
        # Convert names to lowercase before plotting them for proper legend
        for name in  df2_hgh.name:
            df2_hgh['name'] = df2_hgh['name'].str.lower()
        
        my_palette = sns.color_palette(['#206095', '#004662', '#A8BD3A', '#27A0CC', '#118C7B','#F66068'], df2_hgh['name'].nunique())
        sns.lineplot(x = 'year', y = 'rank', data = df2_hgh, hue = 'name', legend=True, ax=ax2,
                       palette=my_palette, linewidth=1.5, marker="o")
        ax2.get_legend().set_title(None)
        ax2.legend(loc = 'lower right')
    
    
    ax2.set_xticks([1996, 1998, 2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018, 2020])
    ax2.set_xticklabels(['1996', '1998', '2000', '2002', '2004', '2006', '2008', '2010', '2012', '2014', '2016', '2018', '2020'])
    
    ax2.set_yticks([1, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
    ax2.set_yticklabels(['1', '100', '200', '300', '400', '500', '600', '700', '800', '900', '1000'])
    
    for ytick in ax2.get_yticks():
        ax2.axhline(y=ytick, color='gray', linewidth=0.5, alpha=0.5)
        
    ax2.invert_yaxis()
    ax2.set(xlabel=None)
    ax2.set(ylabel=None)
    sns.despine(fig2, ax2, top=True, left=True, right=True, offset=10)
    plt.title('Popularity ranking (1 being the most popular)', fontsize=15)
    plt.show()
         
        
interact(f, sex=widgets.ToggleButtons(options=options, description=" "),
            filter=widgets.Text(value='', max_chars=100, continuous_update=False, description="Name:"));

```
